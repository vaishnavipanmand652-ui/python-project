2	Calculate equivalent resistance of resistors connected in series and parallel

```python

r1 = float(input("Enter first resistance (ohms): "))
r2 = float(input("Enter second resistance (ohms): "))


series = r1 + r2

parallel = (r1 * r2) / (r1 + r2)

print("\nEquivalent Resistance:")
print("Series connection:", series, "ohms")
print("Parallel connection:", parallel, "ohms")
```

3	Calculate voltage using Ohm’s Law

```python


current = float(input("Enter current (in amperes): "))
resistance = float(input("Enter resistance (in ohms): "))

voltage = current * resistance

print("\nVoltage =", voltage, "volts")
```

4	Check whether the entered frequency is audio or radio frequency

```python
frequency = float(input("Enter frequency in Hz: "))

if 20 <= frequency <= 20000:
    print("This is an Audio Frequency (AF)")
elif frequency > 20000:
    print("This is a Radio Frequency (RF)")
else:
    print("This frequency is below human hearing range")
```

5	Display various radio frequency bands using if–elif ladder

```python
frequency = float(input("Enter frequency in Hz: "))

if 3e3 <= frequency < 3e6:
    print("Very Low Frequency (VLF)")
elif 3e6 <= frequency < 3e7:
    print("Low Frequency (LF)")
elif 3e7 <= frequency < 3e8:
    print("Medium Frequency (MF)")
elif 3e8 <= frequency < 3e9:
    print("High Frequency (HF)")
elif 3e9 <= frequency < 3e10:
    print("Very High Frequency (VHF)")
elif 3e10 <= frequency < 3e11:
    print("Ultra High Frequency (UHF)")
else:
    print("Frequency out of defined RF bands")
```

6	Display resistor color code using match-case

```python
color = input("Enter resistor color: ").lower()

match color:
    case "black":
        print("Black = 0")
    case "brown":
        print("Brown = 1")
    case "red":
        print("Red = 2")
    case "orange":
        print("Orange = 3")
    case "yellow":
        print("Yellow = 4")
    case "green":
        print("Green = 5")
    case "blue":
        print("Blue = 6")
    case "violet":
        print("Violet = 7")
    case "grey":
        print("Grey = 8")
    case "white":
        print("White = 9")
    case _:
        print("Invalid color entered")
```

7	Demonstrate looping constructs

```python
# For loop example
print("For Loop (1 to 5):")
for i in range(1, 6):
    print(i)

# While loop example
print("\nWhile Loop (1 to 5):")
j = 1
while j <= 5:
    print(j)
    j += 1

# Nested loop example
print("\nNested Loop (Pattern):")
for i in range(1, 4):
    for j in range(1, 4):
        print("*", end=" ")
    print()
```
