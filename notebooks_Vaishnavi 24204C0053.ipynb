2. User-Defined Functions

Develop a Python program using user-defined functions to implement Half Adder and Full Adder logic.


```python
# User-defined function for Half Adder
def half_adder(a, b):
    sum_bit = a ^ b      # XOR
    carry = a & b       # AND
    return sum_bit, carry

def full_adder(a, b, cin):
    sum1, carry1 = half_adder(a, b)
    sum2, carry2 = half_adder(sum1, cin)
    final_carry = carry1 | carry2
    return sum2, final_carry

a = int(input("Enter A (0 or 1): "))
b = int(input("Enter B (0 or 1): "))
cin = int(input("Enter Carry In (0 or 1): "))


ha_sum, ha_carry = half_adder(a, b)

print("\nHalf Adder Output:")
print("Sum =", ha_sum)
print("Carry =", ha_carry)

fa_sum, fa_carry = full_adder(a, b, cin)

print("\nFull Adder Output:")
print("Sum =", fa_sum)
print("Carry =", fa_carry)
```

3. Classes and Objects

Write a Python program to define a Student class with appropriate data members and member functions.
Create objects of the class and display student details.


```python

class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def display_details(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)
        print("------------------")

student1 = Student("Rahul", 101, 85)
student2 = Student("Anita", 102, 90)


student1.display_details()
student2.display_details()
```


5. User-Defined Module

Create and use a user-defined module (for example, a calculator module) to perform arithmetic operations.

```python


def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

x = 20
y = 4

print("Addition:", add(x, y))
print("Subtraction:", subtract(x, y))
print("Multiplication:", multiply(x, y))
print("Division:", divide(x, y))

```


6. NumPy Package

Write a Python program to demonstrate the use of the NumPy package for performing array operations.


```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print("Array a:", a)
print("Array b:", b)

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)

print("Sum of array a:", np.sum(a))
print("Maximum value in array b:", np.max(b))
```
