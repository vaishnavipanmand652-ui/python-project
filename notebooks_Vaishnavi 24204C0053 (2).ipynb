1. List Operations
Develop a Python program to create, access, update, and delete student records using a List data structure.

```python
# Empty list to store student records
students = []

# CREATE (Add students)
students.append(["101", "Amit", 85])
students.append(["102", "Neha", 90])
students.append(["103", "Ravi", 78])

print("Student Records:")
for s in students:
    print(s)

# ACCESS (View one student)
print("\nFirst student record:")
print(students[0])

# UPDATE (Change marks of Ravi)
students[2][2] = 88

print("\nAfter Updating Ravi's marks:")
for s in students:
    print(s)

# DELETE (Remove Neha's record)
students.pop(1)

print("\nAfter Deleting Neha's record:")
for s in students:
    print(s)

```

4. Dictionary Operations
Write a Python program to store and manage student details (Roll No, Name, Marks) using a Dictionary, including create, access, update, delete, and traversal operations.


```python
students = {
    101: {"Name": "Amit", "Marks": 85},
    102: {"Name": "Neha", "Marks": 90},
    103: {"Name": "Ravi", "Marks": 78}
}

print("Student Records:")
print(students)

print("\nDetails of Roll No 102:")
print(students[102])

students[103]["Marks"] = 88

print("\nAfter Updating Ravi's marks:")
print(students)

del students[102]

print("\nAfter Deleting Roll No 102:")
print(students)

print("\nAll Student Details:")
for roll, info in students.items():
    print("Roll No:", roll, "Name:", info["Name"], "Marks:", info["Marks"])
```

6. Laboratory Record Handling
Develop a Python program to manage laboratory resources and availability using Python data structures.

```python
lab_resources = {}

def add_resource():
    name = input("Enter resource name: ")
    status = input("Enter status (Available/In Use): ")
    lab_resources[name] = status
    print("Resource added successfully!\n")

def view_resources():
    if not lab_resources:
        print("No resources found.\n")
    else:
        print("\nLaboratory Resources:")
        for name, status in lab_resources.items():
            print(name, ":", status)
        print()

def update_status():
    name = input("Enter resource name to update: ")
    if name in lab_resources:
        status = input("Enter new status (Available/In Use): ")
        lab_resources[name] = status
        print("Status updated!\n")
    else:
        print("Resource not found!\n")

while True:
    print("1. Add Resource")
    print("2. View Resources")
    print("3. Update Resource Status")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_resource()
    elif choice == "2":
        view_resources()
    elif choice == "3":
        update_status()
    elif choice == "4":
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Try again.\n")
```

7. Data Traversal
Write a Python program to traverse and display all stored student and laboratory records efficiently.



```python
students = {
    101: "Amit",
    102: "Neha",
    103: "Rahul"
}

lab_resources = {
    "Microscope": "Available",
    "Test Tubes": "In Use",
    "Centrifuge": "Available"
}

print("----- Student Records -----")
for roll_no, name in students.items():
    print("Roll No:", roll_no, "| Name:", name)

print("\n----- Laboratory Records -----")
for resource, status in lab_resources.items():
    print("Resource:", resource, "| Status:", status)
```
