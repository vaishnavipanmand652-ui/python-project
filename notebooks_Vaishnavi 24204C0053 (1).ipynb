1. Stack Implementation

Write a Python program to implement a Stack using array with PUSH and POP operations for managing records.



```python
stack = [] 

def push(item):
    stack.append(item)
    print(item, "pushed into stack")


def pop():
    if len(stack) == 0:
        print("Stack is empty! Cannot pop")
    else:
        removed = stack.pop()
        print(removed, "popped from stack")

def display():
    print("Stack:", stack)

push(10)
push(20)
push(30)
display()

pop()
display()

pop()
pop()
pop()  
```

2. Queue Implementation

Develop a Python program to implement a Queue using array with ENQUEUE and DEQUEUE operations for data processing.




```python
queue = []  


def enqueue(item):
    queue.append(item)
    print(item, "added to queue")

def dequeue():
    if len(queue) == 0:
        print("Queue is empty! Cannot remove")
    else:
        removed = queue.pop(0)
        print(removed, "removed from queue")

def display():
    print("Queue:", queue)

enqueue(10)
enqueue(20)
enqueue(30)
display()

dequeue()
display()

dequeue()
dequeue()
dequeue()   
```

3. Singly Linked List

Write a Python program to implement a Singly Linked List with operations such as create, insert, delete, and traverse.




```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def delete(self, key):
        temp = self.head

        if temp and temp.data == key:
            self.head = temp.next
            return

        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next

        if temp is None:
            print("Value not found")
            return

        prev.next = temp.next

    def traverse(self):
        temp = self.head
        print("Linked List:", end=" ")
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
        
sll = SinglyLinkedList()

sll.insert(10)
sll.insert(20)
sll.insert(30)
sll.traverse()

sll.insert(40)
sll.traverse()

sll.delete(20)
sll.traverse()

sll.delete(100)
```

4. Real-Life Application

Design and implement a mini project such as a Student Management System or Employee Management System using Python to perform:

Record creation

Data update

Calculations

Report generation

```python
students = {}
def add_student():
    roll = int(input("Enter Roll No: "))
    name = input("Enter Name: ")
    marks = list(map(int, input("Enter 3 marks (space separated): ").split()))
    students[roll] = {"name": name, "marks": marks}
    print("Student added successfully!\n")

def update_student():
    roll = int(input("Enter Roll No to update: "))
    if roll in students:
        marks = list(map(int, input("Enter new 3 marks: ").split()))
        students[roll]["marks"] = marks
        print("Record updated!\n")
    else:
        print("Student not found!\n")
        
def calculate(roll):
    marks = students[roll]["marks"]
    total = sum(marks)
    avg = total / len(marks)
    return total, avg

def report():
    print("\n----- Student Report -----")
    for roll, data in students.items():
        total, avg = calculate(roll)
        print(f"Roll: {roll}, Name: {data['name']}, Total: {total}, Average: {avg:.2f}")
    print()

while True:
    print("1. Add Student")
    print("2. Update Student Marks")
    print("3. Generate Report")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        add_student()
    elif choice == 2:
        update_student()
    elif choice == 3:
        report()
    elif choice == 4:
        print("Exiting Program...")
        break
    else:
        print("Invalid choice!\n")
```
